## Metadata

First, let's look at the metadata (titles, duration, genre, etc).

```python
# load the dataset
# replace path if different
meta = pd.read_csv("data/MovieSummaries/movie.metadata.tsv"
                   , sep = '\t'
                   , header = None
                  )

# rename the columns
meta.columns = ["movie_id"
                ,"movie_plot_id"
                ,"movie_title"
                ,"movie_release"
                ,"movie_boxoffice"
                ,"movie_duration"
                ,"movie_language"
                ,"movie_country"
                ,"movie_genre"
               ]

print(meta.shape)
meta.head()
```

## Movie Summaries
Second, let's have a look at the summaries.

```python
# load the data into a list
plots = []
# replace path if different
with open("data/MovieSummaries/plot_summaries.txt", 'r', encoding="utf8") as f:
    reader = csv.reader(f, dialect='excel-tab')
    for row in reader:
        plots.append(row)
    
# load it into dataframe
movies = pd.DataFrame(plots)
movies.columns = ["movie_id","movie_plot"]
movies["movie_id"] = movies["movie_id"].astype(int)
movies.head()
```

## Combining Metadata and Movie Summaries

```python
###########################################

# merge dataframes on movie_id
df = pd.merge(movies, meta, on = 'movie_id')
print(df.shape)

###########################################
# The genre is not well formatted. Let's format it into a nice list.

# example
print("Before formatting:", df["movie_genre"][0])

# extract genres
genres = [] 
for i in df['movie_genre']: 
    genres.append(list(json.loads(i).values())) 

# add to dataframe  
df['movie_genre'] = genres

# example
print("After formatting:", df["movie_genre"][0])

###########################################
# filter

# we are only using a few columns
df = df[["movie_id","movie_plot","movie_genre"]]
print(df.shape)

# remove filters without label (i.e. no genre declared)
series = df["movie_genre"].apply(lambda x: len(x))
sup = df.shape[0]
df = df[series>0]
print(sup-df.shape[0], "movies deleted")
print(df.shape)
```

# Preprocessing the Combined Dataset

The basic structure of this tutorial is:
- Clean the text
- Remove stopwords from the cleaned text
- Extract features from the text
- Make predictions
- Return the predicted movie genre tags

## Processing the X

### Visualize the most common words

```python
def freq_words(x, terms = 30, title="Most Frequent Words"): 
    all_words = ' '.join([text for text in x])
    all_words = all_words.split() 
    fdist = nltk.FreqDist(all_words)
    words_df = pd.DataFrame({'word':list(fdist.keys()), 'count':list(fdist.values())}) 
    
    # select the n most frequent words
    d = words_df.nlargest(columns="count", n = terms)
    
    # visualize words and frequencies
    plt.figure(figsize=(12,18))
    ax = sns.barplot(data=d
                     ,x="count"
                     ,y="word"
                     ,color="blue"
                    )
    ax.set(ylabel = 'word')
    plt.title(title,fontsize=18)
    plt.show()

# plot
freq_words(df['movie_plot'], 50, title="Most Frequent Words - Original Text")
```

### Clean the text

```python
# function for text cleaning 
def clean_text(text):
    # remove backslash-apostrophe 
    text = re.sub("\'", "", text) 
    # remove everything except alphabets 
    text = re.sub("[^a-zA-Z]"," ",text) 
    # remove whitespaces 
    text = ' '.join(text.split()) 
    # convert text to lowercase 
    text = text.lower() 
    return text

# print 100 most frequent words: there are many stopwords

df["movie_plot_clean"] = df["movie_plot"].apply(lambda x: clean_text(x))
freq_words(df['movie_plot_clean'], 50, title="Most Frequent Words - Post Cleaning")
```

### Remove Stopwords

On the first time you run *from nltk.corpus import stopwords* you might need to also run *nltk.download('stopwords')*

```python
# function to remove stopwords
def remove_stopwords(text, stop_words):
    no_stopword_text = [w for w in text.split() if not w in stop_words]
    return ' '.join(no_stopword_text)

stop_words = set(stopwords.words('english'))
df['movie_plot_clean'] = df['movie_plot_clean'].apply(lambda x: remove_stopwords(x, stop_words=stop_words))
freq_words(df['movie_plot_clean'], 50, title="Most Frequent Words - Post Cleaning, Stopword Removal")
```

## Processing the Y
### Encode the Genre

```python
with open('utils/decode_genres.json', 'r') as f:
    decode_genres = json.load(f)
    
with open('utils/encode_genres.json', 'r') as f:
    encode_genres = json.load(f)

# encode the dataset

df["movie_genre_encoded"] = df["movie_genre"].apply(lambda x: set([encode_genres[i] for i in x]))
```
