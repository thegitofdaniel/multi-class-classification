As each movie may have one genre ore more, this is multi-label classification (i.e. each genre is a label, and a movie may have more than one genre).

- One technicality:
    - Multi-class is choosing one exclusive category out of many;
    - Multi-label is choosing at least one category out of many.

## Formatting

```python
###########################################
# The genre is not well formatted. Let's format it into a nice list.

# example
print("Before formatting:", meta["movie_genre"][0])

# extract genres
genres = [] 
for i in meta['movie_genre']: 
    genres.append(list(json.loads(i).values())) 

# add to dataframe  
meta['movie_genre'] = genres

# example
print("After formatting:", meta["movie_genre"][0])
```

## How are genres distributed?

A quick look at the list of genres shows they are not too well behaved.

Some examples:

Some genres should be logically derived, but I wonder if they always are:
- Logical combinations
    - "Action", "Adventure" -> Action-Adventure
    - "Comedy", "Horror" -> Horror-Comedy
- General and specific
    - "Children's"
    - "Children's Entertainment"
    - "Children's Fantasy"
    - "Children's Issues"
    - "Children's/Family"

Some genres are 'wrong':
- They should be the same:
    - "Anti-war" and "Anti-war film"
    - Others that are not so clear cut ("Political cinema" and "Political drama")
- Arguably not genre:
    - "Bollywood"
    - "The Netherlands in World War II"
    - "War effort"
    - "Black-and-white"

Also:
- There are some genres with very just a few (if not zero or only one) movies.

Conclusions:
- There are too many categories.
- They should be cleaned (e.g. "Anti-war" and "Anti-war film" -> "Anti-war").
- Logical rules should be applied to sanitize the dataset (e.g. IF "Action" and "Comedy" THEN "Action-Commedy").
- Categories with very few movies should be rethought and possibly eliminated.

# How are they correlated?

The graph is not very clear because there are too many categories (i.e. sparse matrix). However, we can take a few points:
- There are (maybe) too many categories.
- Most genres are not correlated (positively or negatively) with each other.
- There are some bumps in the diagonal, suggesting similar category names are marked together quite often (e.g. subtypes).
- There are a few islands of correlation.

```python
# let's create a sparse matrix
decode_genres = dict([i,unique_genres[i]] for i in range(len(unique_genres)))
encode_genres = dict([unique_genres[i],i] for i in range(len(unique_genres)))
meta["movie_genre_encoded"] = meta["movie_genre"].apply(lambda x: set([encode_genres[i] for i in x]))

# save these encoders/decoders for later
with open('../utils/decode_genres.json', 'w') as f:
    json.dump(decode_genres, f)
    
with open('../utils/encode_genres.json', 'w') as f:
    json.dump(encode_genres, f)

# encode the dataset
warnings.filterwarnings("ignore")
df = pd.DataFrame()
for i in decode_genres:
    df[i] = meta["movie_genre_encoded"].apply(lambda x: i in x)
warnings.filterwarnings("default")

# altenatively:
#multilabel_binarizer = MultiLabelBinarizer()
#multilabel_binarizer.fit(df['movie_genre_encoded'])
#y = multilabel_binarizer.transform(df['movie_genre_encoded'])
#df = pd.DataFrame(y)

print(df.columns)
```

```python
# calculate the correlations
corr_df = df.corr(method='pearson')

# create a custom color map
cdict = {'red':  ((0.0, 0.0, 0.0),   # no red at 0
                  (0.5, 1.0, 1.0),   # all channels set to 1.0 at 0.5 to create white
                  (1.0, 0.8, 0.8)),  # set to 0.8 so its not too bright at 1

        'green': ((0.0, 0.8, 0.8),   # set to 0.8 so its not too bright at 0
                  (0.5, 1.0, 1.0),   # all channels set to 1.0 at 0.5 to create white
                  (1.0, 0.0, 0.0)),  # no green at 1

        'blue':  ((0.0, 0.0, 0.0),   # no blue at 0
                  (0.5, 1.0, 1.0),   # all channels set to 1.0 at 0.5 to create white
                  (1.0, 0.0, 0.0))   # no blue at 1
       }
GnRd = LinearSegmentedColormap('GnRd', cdict).reversed()

# Make a figure and axes
fig, ax = plt.subplots(1,figsize=(24,24))

# Plot the fake data
p=ax.pcolormesh(corr_df
                ,cmap=GnRd
                ,vmin=-1
                ,vmax=1
               )

# Make a colorbar
fig.colorbar(p, ax=ax)

plt.show()
```
